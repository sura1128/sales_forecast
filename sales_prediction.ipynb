```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
store_sales = pd.read_csv("train.csv")
store_sales.head(10)
```

Check for null values

```python
store_sales.info()
```

Droopping store and item columns

```python
store_sales = store_sales.drop(['store', 'item'], axis=1)
```

```python
store_sales.info()
```

Converting date from object dataType to dateTime datatype

```python
store_sales['date'] = pd.to_datetime(store_sales['date'])
```

```python
store_sales.info()
```

Converting date to a month period and sum the number of items in each month

```python
store_sales['date'] = store_sales['date'].dt.to_period("M")
monthly_sales = store_sales.groupby('date').sum().reset_index()
```

Convert the resulting date column to timestamp datatype

```python
monthly_sales['date'] = monthly_sales['date'].dt.to_timestamp()
```

```python
monthly_sales.head(10)
```

Visualize monthly sales

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Monthly Customer Sales")
plt.show()
```

Call the difference on the sales columns to make the sales data stationary (check the difference with previous sale) - this drops the value in the first row. We are calculating the increase or decrease in sales.

```python
monthly_sales['sales_diff'] = monthly_sales['sales'].diff()
monthly_sales = monthly_sales.dropna()
monthly_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['date'], monthly_sales['sales_diff'])
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Monthly Customer Sales Difference")
plt.show()
```

Drop the dates and sales as we will only deal with stationary sale data.

```python
supervised_data = monthly_sales.drop(['date', 'sales'], axis=1)
supervised_data.head(10)
```

Preparing the supervised data by adding columns based on months (every 12 months)

```python
for i in range(1,13):
  col_name = 'month_' + str(i)
  supervised_data[col_name] = supervised_data['sales_diff'].shift(i)
supervised_data = supervised_data.dropna().reset_index(drop=True)
supervised_data.head(10)
```

Split data into train and test datasets

```python
train_data = supervised_data[:-12] # previous 12 months
test_data = supervised_data[-12:] # coming 12 months
print ("Train Data Shape: ", train_data.shape)
print ("Test Data Shape: ", test_data.shape)
```

Scale data to a min/max range for the datasets

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
scaler.fit(train_data)
train_data = scaler.transform(train_data)
test_data = scaler.transform(test_data)
```

```python
x_train, y_train = train_data[:,1:], train_data[:,0:1]
x_test, y_test = test_data[:,1:], test_data[:,0:1]
y_train = y_train.ravel() # flattens the array, first column is always the output
y_test = y_test.ravel()
print("x_train shape: ", x_train.shape)
print("y_train shape: ", y_train.shape)
print("x_test shape: ", x_test.shape)
print("y_test shape: ", y_test.shape)
```

Make a prediction data frame

```python
sales_dates = monthly_sales['date'][-12:].reset_index(drop=True)
predict_df = pd.DataFrame(sales_dates)
print (predict_df)
```

```python
act_sales = monthly_sales['sales'][-13:].to_list()
print (act_sales)
```

Create the linear regression model and the predicted output

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
lr_pre = lr_model.predict(x_test)
```

```python
lr_pre = lr_pre.reshape(-1,1)
# Set matrix that contains input features of test data and also the predicted output
lr_pre_test_set = np.concatenate([lr_pre, x_test], axis=1) # concatenate both of them
lr_pre_test_set = scaler.inverse_transform(lr_pre_test_set)
result_list = []
for index in range(0,len(lr_pre_test_set)):
  result_list.append(lr_pre_test_set[index][0] + act_sales[index])
lr_pre_series = pd.Series(result_list, name="Linear Sales Prediction")
predict_df = predict_df.merge(lr_pre_series, left_index=True, right_index=True)
```

```python
lr_mse = np.sqrt(mean_squared_error(predict_df['Linear Sales Prediction'], monthly_sales['sales'][-12:]))
lr_mae = mean_absolute_error(predict_df['Linear Sales Prediction'],  monthly_sales['sales'][-12:])
lr_r2 = r2_score(predict_df['Linear Sales Prediction'], monthly_sales['sales'][-12:])

print("Linear Regression MSE: ", lr_mse)
print("Linear Regression MAE: ", lr_mae)
print("Linear Regression R2: ", lr_r2)
```

Visualization of prediction against actual sales

```python
plt.figure(figsize=(15,5))
# actual sales
plt.plot(monthly_sales['date'], monthly_sales['sales'])
# predicted sales
plt.plot(predict_df['date'], predict_df['Linear Sales Prediction'])
plt.title("Sales forecast using Linear Regression Model")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend(['Actual Sales', 'Predicted Sales'])
plt.show()
```
